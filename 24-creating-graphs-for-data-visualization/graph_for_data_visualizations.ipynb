<h3>Load Data from Amazon S3</h3>

```python
df = pd.read_csv('s3://knodax-ml-datasets/customer_sales_data.csv')
df.head()
```

<h3>Generate a Scatter Plot</h3>
Visualize relationships between two continuous variables:

```python
sns.scatterplot(data=df, x='age', y='income')
plt.title('Scatter Plot: Age vs Income')
plt.show()
```

<h3>Create a Time Series Plot</h3>
Convert the date column to datetime if needed and plot:

```python
df['date'] = pd.to_datetime(df['date'])
sns.lineplot(data=df, x='date', y='sales')
plt.title('Time Series Plot: Date vs Sales')

# This rotates the x-axis tick labels by 45 degrees. It's useful when the labels are too long or too close together, which can make them hard to read if they are all horizontal.
plt.xticks(rotation=45)

# Automatically adjusts the subplot parameters (margins and spacing) so that the content (like labels and titles) fits within the figure area neatly. This prevents labels from getting cut off, especially when you've rotated them.
plt.tight_layout()

plt.show()
```

<h3>Draw a Histogram</h3>
Examine the distribution of a numeric variable:

```python
sns.histplot(df['income'], bins=20, kde=True)
plt.title('Histogram of Income')
plt.show()
```

<h3>Generate a Box Plot</h3>
Visualize data spread and outliers for groups:

```python
sns.boxplot(data=df, x='category', y='income')
plt.title('Box Plot: Income by Category')
plt.show()
```

<h3>Summarize insights</h3>
	Scatter plot shows correlation. <br>
	Time series reveals trends or seasonality.<br>
	Histogram highlights skewness or normality.<br>
	Box plot detects outliers and group-wise variations.
